```python
card_images = df[(pd.notnull(df.cost))&(pd.notnull(df.img))]
```

```python
images = {}
for card, image in card_images.img.items():
    images[card] = image
```

```python
def save_image(card, image_url):
    with open('images/%s.jpg' % card, 'wb') as handle:
            response = requests.get(image_url, stream=True)

            if not response.ok:
                print(response)

            for block in response.iter_content(1024):
                if not block:
                    break

                handle.write(block)
for card, image in images.items(): 
    save_image(card, image)
```
